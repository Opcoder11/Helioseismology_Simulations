```python
from sunpy.net import Fido, attrs as a
from astropy import units as u

result = Fido.search(
    a.Time('2016-06-16T07:21:00', '2016-06-16T07:21:30'),
    a.Instrument("aia"),
    a.Wavelength(1700 * u.angstrom),
    a.Sample(12 * u.second)  # ✅ this avoids JSOC timeouts
)

print(result)

downloaded_files = Fido.fetch(result,max_conn=1,progress=True)
print(downloaded_files)
```
